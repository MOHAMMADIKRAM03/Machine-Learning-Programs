```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
file_path = 'C:\\Users\\ASUS\\PycharmProjects\\P1\\P1\\salary_data.csv'
salary_df = pd.read_csv(file_path)
X = salary_df.iloc[:, 0].values.reshape(-1, 1) 
y = salary_df.iloc[:, 1].values  
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred = lin_reg.predict(X)
plt.scatter(X, y, color='red', label='Data points')
plt.plot(X, y_pred, color='green', label='Regression line')
plt.title('Simple Linear Regression')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```
